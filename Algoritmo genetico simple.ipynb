```python
import random
import numpy as np


ciclos = 4
# funcion que solo genera binarios al azar
def GeneraBinarios():
    lista = []
    for a in range(9):
        lista.append(random.randint (0, 2))
    return lista

#funcion que divide entero y decimal y calcula si esta en el rango de X, 0<= X < 3.14
def EnteroDecimal(binario):
    entero = 0
    decimal = 0
    #suma toda l parte entera
    cont = 2
    for a in range(cont):
        entero = entero + (binario[a]*cont)
        cont = cont-1
    #suma toda la parte decimal   
    cont = 6
    for a in range(2,cont+3):
        decimal = decimal + (np.power(2,cont)*binario[a])
        cont=cont-1
    # evalua si esta en el rango que es como maximo 3.14
    while entero == 3:
        if decimal > 14:
            gen = GeneraBinarios()
            return EnteroDecimal(gen)
        else:
            return entero, decimal, binario
    return entero, decimal, binario

def ValoresOptimos(poblacion):
    
    for a in range(poblacion):
        gen = GeneraBinarios()
        entero,decimal, bina = EnteroDecimal(gen)
        matriz.append(bina)
        
def MostrarMatriz(poblacion):
    print("entero, decimal,     binario")
    for a in range(poblacion):
        entero,decimal, bina = EnteroDecimal(matriz[a])
        print("  ",entero,",",decimal,"     ", bina)
        
def OrdenarSegunFitness(poblacion):
    #ordenando segun fitnes en este caso segun el menos
    for a in range(poblacion):
        for b in range(poblacion):
            if EnteroDecimal(matriz[a])<EnteroDecimal(matriz[b]):
                temp = matriz[a]
                matriz[a]=matriz[b]
                matriz[b]=temp
                
def EnteroDecimal2(binario):
    entero = 0
    decimal = 0
    #suma toda l parte entera
    cont = 2
    for a in range(cont):
        entero = entero + (binario[a]*cont)
        cont = cont-1
    #suma toda la parte decimal   
    cont = 6
    for a in range(2,cont+3):
        decimal = decimal + (np.power(2,cont)*binario[a])
        cont=cont-1
    return entero, decimal, binario

def aleatorio():
    return random.randint(1,9)

def cruce(matriz,ubicacionCorte):
    gen1 = matriz[0]
    gen2 = matriz[1]
    listaGen1 = []
    listaGen2 = []
    for a in range(9):
        if a<ubicacionCorte:
            listaGen1.append(gen1[a])
            listaGen2.append(gen2[a])
            print()
        else:
            listaGen2.append(gen1[a])
            listaGen1.append(gen2[a])
    matriz[0] =listaGen1
    matriz[1] =listaGen2
    
    return matriz

def mostrarGenCortado(matriz,ubicacionCorte):
    tem1 =[]
    tem2 =[]
    tem11 =[]
    tem22 =[]
    for a in range(9):
        if a <ubicacionCorte:
            tem1.append(matriz[0][a])
            tem11.append(matriz[1][a])
        else:
            tem2.append(matriz[1][a])
            tem22.append(matriz[0][a])
    print(tem1," ", tem2,"     ",tem11," ",tem22)
    print("se realiza el intercambio de genes")
    print(tem1," ", tem22,"     ",tem1," ",tem22)
    print ("hacer el cruce")
    
    
def ProvabMutar():
    return random.randint(0,2)

def QuienesMutaron(matriz):
    for a in range(4):
        if ProvabMutar():
            temp = random.randint(0,9)
            if matriz[a][temp] == 1:
                matriz[a][temp]=0
            else:
                matriz[a][temp]=1
            print("gen Nro: ",a+1,"  muto poscicion: ",temp+1)
    return matriz
```

```python
#almacenando en la matriz solo los valores en el rango indicado
matriz = [] 
poblacion = 4
gen = GeneraBinarios() 
EnteroDecimal(gen)
ValoresOptimos(poblacion)
print("Muestar valores optimos en el rango de 0 a 3.14")
MostrarMatriz(poblacion)
OrdenarSegunFitness(poblacion)
print("Muestar valores ordenado para evaluarlos")
MostrarMatriz(poblacion)
```
